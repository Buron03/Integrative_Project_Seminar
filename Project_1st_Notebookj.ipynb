# ◢ Ecuaciones Diferenciales

## 1.1. Introducción

icon.png

<p>Una ecuación diferencial es una ecuación matemática que relaciona una función con sus derivadas. En las matemáticas aplicadas, las funciones usualmente representan cantidades físicas, las derivadas representan sus razones de cambio, y la ecuación define la relación entre ellas. Como estas relaciones son muy comunes, las ecuaciones diferenciales juegan un rol primordial en diversas disciplinas, incluyendo la ingeniería, la física, la química, la economía, y la biología.</p>

<p>En las matemáticas puras, las ecuaciones diferenciales se estudian desde perspectivas diferentes, la mayoría concernientes al conjunto de las soluciones de las funciones que satisfacen la ecuación. Solo las ecuaciones diferenciales más simples se pueden resolver mediante fórmulas explícitas; sin embargo, se pueden determinar algunas propiedades de las soluciones de una cierta ecuación diferencial sin hallar su forma exacta.</p>

# ◢ Solución y Pasos

## 2.1. Definiciones y Pasos Elementales

<p>En general, decir que un <strong>paso</strong> es una asociación binaria (entre dos objetos) de una expresión algebraica (normalemente una ecuación) y una leyenda que le describe. Definimos además una <strong>solución</strong> como una colección finita de pasos. Dada una ecuación diferencial, el programa retornará una solución en caso de que pueda encontrarse. Dicha solución es la misma para una misma ecuación diferencial.</p><br>

<p>Podemos entonces decir que para construir un paso se requiere de una manipulación algebraica de una expresión matemática previamente dada. La idea es que partiendo de la expresión dada por el usuario se puedan construir los pasos que compondrán a la solución. Siguiendo esta idea, podemos decir que es posible construir todos los pasos por medio de manipulación algebracias definidas. Dedicaremos el resto de este apartado a describir cada uno de los pasos que se pueden presentar y a asignarles un nivel de dificultad asociado al desarrollo del mismo a nivel informático.</p><br>

<p>Podemos separar los pasos más elementales en dos grupos: aquellos que se operan sobre una igualdad y aquellos que no lo hacen. Siendo rigurosos con esta idea, podemos definir al operador $\gamma$ como una manipulación algebraica. Entonces tenemos que los dos pasos son como siguen:</p><br>

<p>Paso sobre la igualdad: $f = g \Rightarrow \gamma[f] = \gamma[g]$</p>
<p>Paso sin la igualdad: $f \Rightarrow \gamma[f]$</p>
<p>Donde $f, g$ son expresiones algebraicas cualesquiera</p><br>

<p>A aquellas pasos que se operan sobre una igualdad les llamaremos <strong>operaciones algebraicas</strong>, mientras que a los pasos que no se aplican sobre una igualdad serán conocidos como <strong>transformaciones algebraicas</strong>. En general, tenemos que para construir una operación o transformación algebraica se utilizarán métodos de matemáticas simbólicas concretos, simplificados y sin intermediarios. Para trabajar con estos pasos se utilizarán los métodos en SymPy de manipulación algebraica. Las manipulación algebraicas que se utilizarán con su nivel de dificultad son:</p><br>

> <strong>Suma</strong>

<p>Operación: $f+h = g+h$</p>
<p>Transformación: $f \leftarrow f+h$</p>
<p><strong>Dificultad: 1</strong></p><br><br>

> <strong>Resta</strong>

<p>Operación: $f-h = g-h$</p>
<p>Transformación: $f \leftarrow f-h$</p>
<p><strong>Dificultad: 1</strong></p><br><br>

> <strong>Multiplicación</strong>

<p>Operación: $f*h = g*h$</p>
<p>Transformación: $f \leftarrow f*h$</p>
<p><strong>Dificultad: 1</strong></p><br><br>

> <strong>Division</strong>

<p>Operación: $f/h = g/h$</p>
<p>Transformación: $f \leftarrow f/h$</p>
<p><strong>Dificultad: 1</strong></p><br><br>

> <strong>Composición</strong>

<p>Operación: $h(f) = h(g)$</p>
<p>Transformación: $f \leftarrow h(f)$</p>
<p><strong>Dificultad: 1</strong></p><br><br>

<p>(Note que aplicar un paso sobre la igualdad es equivalente a realizar un paso sin igualdad para cada lado de la igualdad, por esto se podría considerar que el paso mínimo es la transformación; no obstante, hablamos de hablamos del paso de la igualdad como si fuera atómico por practicidad en la construcción de algoritmos con los que trabajeremos más adelante)</p><br>

<p>Por medio de combinaciones de estos pasos elementales se pueden construir pasos más complejos, los cuáles consideraemos como los pasos más pequeños a mostrar en la solución con la intención de no realizar una exposición demasiado extensa de la solución y perder detalle en las ideas clave de la solución. Los <strong>pasos compuestos</strong> son aquellos que pueden ser descritos como una combinación finita de pasos elementales; poseen un propósito específico y representan la idea completa de un paso. Estso pasos son los que aparecerán en la exposición final de la solución, y pueden separarse en dos grupos dependiendo del propósito con el que se lleven a cabo: algebraicos e integrales. </p><br>

## 2.2. Pasos Algebraicos

<p>Se definen a un <strong>paso algebraico</strong> como una combinación de pasos elementales que realizan un trabajo meramente algebraico sobre la expresión. Dichos pasos tienen fines de manipulación y ordenamiento de la ecuación. Pese a que la cantidad de pasos elementales que componen a un paso algebraico es finita, no se puede conocer con certeza la cantidad exacta de ellos que se requiere para llevar a cabo el paso en cuestión, por lo que su dificultad no generalizarse a un valor estándar. No obstante, con base a las ecuaciones más frecuentes con las que se pretende trabajar, se puede realizar una estimación de la cantidad de pasos elementales que le tomará; con la intención de no abondanar la idea de que es una aproximación, la dificultad se presenta en forma de rango. Los posibles pasos algebraicos que se podrán presentar son:</p><br>

> <strong>Sustitución</strong>

<p>Reemplaza todas las ocurrencias de un término por una nueva expresión</p>

## 2.3. Pasos Integrales

### 2.3.1. Generalidades

<p>Se define a un <strong>paso integral</strong> como una combinación de pasos elementales que desempeñan la integración de una función. Un paso integral comienza con la expresión explícita de la integral y concluye con el resultado de dicha integral en caso de tener una solución. Si la integral no posee solución matemática conocida, entonces el paso integral regresa un mensaje indicando que no se tiene solución simbólica para dicha integral junto con la integral en su forma explícita. Debido a la gran variedad de integrales que pueden encontrarse, se plantea definir dos tipos de pasos integrales en función del papel que desempeñan dentro de la solución de la integral: Un <strong>paso integral recursivo</strong> será aquel que no resuelve la integral pero genera nuevas integrales que pueden resolverse posteriormente; mientras que un <strong>paso integral atómico</strong> será aquel que resulve directamente la integral sin intermediarios. Se busca entonces que todo paso integral puede expresarse como una cantidad finita de pasos integrales atómicos por medio de una cantidad finita de pasos integrales recursivos.</p>

La dificultad de un paso.

Los pasos atómicos poseen una dificultad establecida que está asociada con su nivel de detección (los carácteres que son necesarios verificar para comprobar que se trata de ese paso atómico en concreto y la clasificación previa dentro del tipo de pasos atómicos disponibles); al igual que los pasos recursivos

Los pasos compuestos poseen una dificultad igual a la suma de las dificultades de los pasos atómicos que lo componen y de las dificultades de los pasos recursivos necesarios para llegar a las expresiones atómicas

Existe un acumulador de dificultad que controla el límite de pasos recursivos y atómicos que pueden emplearse para llevar a cabo un paso. Este acumulador se conoce como pila de dificultad, la cuál consiste en un registro sobre el cual se va sumando la dificultad de cada uno de los pasos atómicos o recursivos que se han realizado a lo largo del paso. La pila de dificultad posee un límite finito, por lo que la cantidad de pasos recursivos y atómicos que podrán realizarse es igualmente finito. La pila de dificultad va acompañada con un par de banderas de control del paso: bandera de chequeo y bandera de sobreflujo. La bandera de chequeo comprueba si hace falta hacer algo más en la expresión para terminar el paso (integrar alguna expresión, simplificar un término, etc.); mientras que la bandera de sobreflujo nos indica si la pila de dificultad ha alcanzado su límite y ha provocado un sobreflujo. Ambas banderas se inicializan antes de cada paso y son revisadas durante cada iteración del mismo. Si la bandera de sobreflujo es activada, entonces significa que el paso es demasiado complejo para su cómputo y por ello se regresa un paso nulo como respuesta al método, el cual consiste en la misma expresión algebraica que llegó junto con la invocación y una leyendo que indica que el paso no pudo llevarse a cabo por su dificultad. Dentro de la solución general (colección de pasos compuestos) se quedaría trunco el desarrollo en caso de que un paso tuviera está composición. Si la bandera de chequeo es activada significa que la expresión termino de procesarse lo que indica que se terminó el paso. La manera de verificar esto es observar que cada uno de los pasos que quedan para terminar son atómicos. 

Por lo anterior, la estrategia general para llevar a cabo un paso compuesto consiste en:

1. Verificar atomicidad de la expresión mediante la consulta en el catálogo de pasos atómicos correspondiente

2. Verificar si alguno de los pasos recursivos elementales puede ser aplicado 

3. Verificar si alguno de los pasos recursivos no elementales puede ser aplicado 

Si se encuentra que la expresión es atómica entonces habremos terminado está iteración (que podría ser incluso la iteración completa del paso compuesto; esto es,  que el paso inicial sea atómico). Al verificarlo, se regresa a la interior iteración con la intención de verificar si quedaron operaciones recursivas por resolver; en dado caso se procede a invocar nuevos pasos sobre estas expresiones para continuar con la búsqueda de las expresiones atómicas.

<p>Una ecuación diferencial de primer orden es <strong>separable</strong> cuando puede llegarse a la siguiente forma por medios algebraicos convencionales:</p><br>

<p> Llegamos a su solución por medio de integración directa por ambos lados de la ecuación, esto es:</p><br>

$\int_{}^{}p(x)dx=\int_{}^{}q(y)dy$
 
<p>Si además $P(x)=\int_{}^{}p(x)dx$, y que $Q(y)=\int_{}^{}q(y)dy$, la ecuación anterior puede verse como:<p>

$P(x) = Q(y)$<br><br>

<p>Esta última expresión se conoce como <strong>solución implíctia</strong>, ya que no da una definición explícita para $y(x)$. Despejar $y$ de la ecuación anterior nos genera la <strong>solución explícita</strong> de la ecuación diferencial</p>

>>> <h3>2.1.1.3. Ecuaciones Diferenciales Lineales</h3>



<p>Una ecuación diferencial de primer orden es <strong>lineal</strong> cuando puede llegarse a la siguiente forma por medios algebraicos convencionales:
<br>$\frac{dy}{dx} = f(x, y) = p(x)y + q(x)$</p><br>

<p>Utilizando un cambio de variable de tal manera que:
<br>$v(x)\frac{dy}{dx} - p(x)v(x)y= \frac{d(v(x)y)}{dx}$; de modo que $v$ sea también función de $x$ tal que $v(x)>0$ para toda $x$<p><br>

<p>Si multiplicamos ambos lados de la igualdad por $v$, se tiene que:
<br>$v(x)\frac{dy}{dx} - p(x)v(x)y = q(x)v(x)$<br><br>
$\frac{d(v(x)y)}{dx} = q(x)v(x)$<br><br>
$d(v(x)y) = v(x)q(x)dx$<br><br>
$v(x)y = \int{}{}v(x)q(x)dx$<br><br>
$y = \frac{1}{v(x)}\int{}{}v(x)q(x)dx$</p><br>

<p>A este factor $v$ se le conoce como <strong>factor de integración</strong>. Para obtenerlo, veamos la condición con la que se impuso el factor:<br><br>
$\frac{d(v(x)y)}{dx} = v(x)\frac{dy}{dx} - p(x)v(x)y$<br><br>
$v(x)\frac{dy}{dx} + y\frac{dv}{dx} = v(x)\frac{dy}{dx} - p(x)v(x)y$<br><br>
$y\frac{dv}{dx} = -p(x)v(x)y$ <br><br>

Si $y$ es diferente a $0$ para todo $x$, entonces se tiene que la igualdad se da cuando:<br><br>
$\frac{dv}{dx} = -p(x)v(x)$<br><br>
$\frac{dv}{v(x)} = -p(x)dx$<br><br>
$\int{}{}\frac{dv}{v(x)} = -\int{}{}p(x)dx$<br><br>
$\ln{v(x)} = -\int{}{}p(x)dx$<br><br>
$v(x) = e^{-\int{}{}p(x)dx}$<br><br>
</p>


>>><h3>Ecuaciones Reducibles a la forma Linear <h3>

<p>Una ecuacion de la forma </p>
$\frac{dy}{dx} + Py = Qy^n$, 
<p> Donde $P$ y $Q$ son funciones de x, puede volverse linear por un cambio de variable. Dividiendo por $y^n$, se vuelve</p>
$y^{-n}\frac{dy}{dx}+Py^{-n+1} = Q$
<p>Si tomamos</p>
$y^{1-n} = u$
<p>Como una nueva variable, la ecuacion toma la forma</p>
$\frac{1}{1-n}\frac{du}{dx} +Pu =Q,$ <br>
<p>La cual es linear</p><br>

<p>Por ejemplo, considere la ecuacion </p>

$\frac{dy}{dx} + \frac{2}{x}y^{-2} = \frac{y^3}{x^3}$
<p>Sea $u = y^{-2}$
<p>Entonces</p>
$\frac{du}{dx} = -2y^{-3}\frac{dy}{dx}$<br>
<p>luego</p>
$y^{-3}\frac{dy}{dx} = -\frac{1}{2}\frac{du}{dx}$
<p>Sustituyendo estos valores, obtenemos</p>
$-\frac{1}{2}\frac{du}{dx} + \frac{2}{x}u = \frac{1}{x^3}$
<p>y por tanto </p>
$\frac{du}{dx} - \frac{4}{x}u = -\frac{2}{x^3}$<br>
<p>La cual es una ecuacion <strong>linear</strong> </p>

>>> <h3>2.1.1.4. Ecuaciones Diferenciales Exactas</h3>


<p>Una ecuación </p>
${du=0}$
<p>Obtenida al igualar a 0 el diferencial total de una funcion ${u}$ de ${x}$ y ${y}$ es llamada ecuacion diferencial <strong>exacta</strong>.</p>
<p>La solucion de este tipo de ecuacion es</p>
${u = c.}$<br>

<p>La condicion necesaria para que ${Mdx + Ndy}$ sea una ecuacion diferencial exacta es

$\frac{∂M}{∂y} = \frac{∂N}{∂x} $<br>
<p>Lo cual es equivalente a que</p>
${Mdx + Ndy = 0}$
<p>Sea una ecuacion diferencial exacta</p><br><br>

<p>Una ecuacion diferencial exacta con frecuencia puede ser resuelta por inspeccion. Para encontrar ${u}$ es necesario obtener una funcion cuyo diferencial total sea</p>
${du = Mdx + Ndy}$
<p>y entonces</p>
$\frac{∂u}{∂x} ={M}.$

<p>Integrando con ${y}$ como constante, obtenemos</p>
$u = \int_{}^{}Mdx + f(y)$ <br><br>


<p> Puesto que $y$ es constante en la integracion, la constante de integracion puede ser una funcion de $y$. Esta funcion puede ser encontrada igualando el total diferencial de $u$ a $Mdx + Ndy$.</p>
<p>Dado que $df(y)$ solo puede dar terminos que contienen a $y$, usualmente $f(y)$ se puede encontrar integrando los terminos en $Ndy$ que no contienen a $x$</p>

<p>En casos excepcionales este proceso puede que <strong>no</strong> de el resultado correcto, por tanto la respuesta debe ser <strong>comprobada</strong> por derivacion.</p>




>>>><h3>2.1.1.4.1  Factores de Integracion</h3>

<p>Si una ecuacion de la forma $Mdx + Ndy = 0$ no es exacta puede volverse exacta al multiplicarse por factor apropiado. Dicho multiplicador es llamado <strong>Factor de integracion</strong>.</p>

<p>Por ejemplo, la ecuacion </p>
$xdy - ydx = 0$
<p>no es exacta. Pero si es multiplicada por $\frac{1}{x^2}$, toma la forma: </p>
$\frac{xdy-ydx}{x^2} = d(\frac{y}{x}) = 0$
<p>que es exacta. Tambien se vuelve exacta cuando es multiplicada por $\frac{1}{y^2}, \frac{1}{xy}$.</p><br><br>

<p>A pesar de que la ecuacion de la forma $Mdx + Ndy = 0$ siempre tiene factores integrantes, <strong>no existe</strong> ningun metodo general para obtenerlos</p>




>>> <h3>2.1.1.4. Ecuaciones Diferenciales Homogéneas</h3>



<p>Una funcion $f(x,y)$ es llamada homogenea de grado $n$ si:</p>

$f(tx, ty) = t^nf(x,y)$<br>

<p>Por tanto $\sqrt{x^2+y^2}$ es homogenea de primer grado; puesto que </p>
$\sqrt{x^2t^2+y^2t^2}=t\sqrt{x^2+y^2}$<br>

<p>Es facil comprobar que un polinomio cuyos terminos sean todos de grado $n$ es una funcion homogenea de grado $n$.</p> 
<p>La ecuacion diferencial </p>
$Mdx + Ndy = 0$
<p>Es llamada homogenea si $M$ y $N$ son funciones homogeneas del mismo grado. Para resolver una ecuacion homogenea sustituya: </p>
$y = vx$
<p>La nueva ecuacion diferencial sera <strong>separable </strong> </p><br><br>

<p>Por ejemplo, considere </p>
$x\frac{dy}{dx} - y = \sqrt{x^2 + y^2}$
<p>Esta es una ecuacion homogenea de primer grado. Sustituyendo $y = vx$, se vuelve</p>

$x(v + x\frac{dv}{dx}) -vx = \sqrt{x^2 + v^2x^2}$ <br>
<p>y por tanto </p>
$x\frac{dv}{dx} = \sqrt{1+v^2}$<br>
<p> La cual es una ecuacion diferencial separable, con solucion:</p>
$x = c(v + \sqrt{1+v^2})$<br>
<p>Remplazando $v$ por $\frac{y}{x}$ y resolviendo, la solucion se vuelve: </p>
$x^2 - 2cy = c^2$

>><h3>Cambio de Variable</h3>
<p>Se resolvieron las ecuaciones homogeneas tomando como nueva variable $v = \frac{y}{x}$</p><br>
<p> Puede que sea posible reducir cualquier ecuacion a una forma mas simple tomando alguna funcion $u$ de $x$ y $y$ como nueva variable o tomando dos funciones $u$ y $v$ como nuevas variables. Dichas funciones son normalmente sugeridas por la ecuacion.</p><br>

<p>Por ejemplo, considere</p>
$(x-y)^2\frac{dy}{dx} = a^2$<br>
<p> Sea $x-y = u$ entonces</p>
$1-\frac{dy}{dx} = \frac{du}{dx}$<br>
<p>y la ecuacion diferencial se vuelve</p>
$u^2(1-\frac{du}{dx}) = a^2,$<br>
<p>Luego</p>
$u^2 - a^2 = u^2 \frac{du}{dx}$<br>
<p>Las variables son separables ahora. Su solucion es:</p>
$x = u +\frac{a}{2}ln\frac{u-a}{u+a}+c = \frac{a}{2}ln\frac{x-y-a}{x-y+a}+c$







>>> <h3>2.1.1.5. Ecuaciones Diferenciales Lineares: Consideraciones Adcionales</h3>






<p>La transformada de Laplace puede ser utilizada para resolver ecuaciones diferenciales lineares </p>

<p>El metodo funciona de la siguiente manera: Estamos buscando una funcion $y(t)$ como solucion de la ecuacion diferencial. Dada la convencion utilizada dicha funcion es $Y(s)$ al aplicarse la transformada de Laplace. </p>

<p>1.- Aplicar la transformada de Laplace</p>
<p>2.- Resolver la ecuacion algebraica en $Y(s)$. Esto significa resolver la ecuacion en el dominio de $Y(s)$, donde es sencilla de resolver. El resultado es $Y(s)=\frac{p(s)}{q(s)}$, donde $p(s)$ y $q(s)$ son polinomios.</p>
<p>3.- La transformada inversa (esta es la parte mas compleja)</p><br>

<p>Una de las propiedades mas importantes de la transformada de Laplace se expresa a continuacion. Esta propiedad es la que nos permite tratar una <strong>ED</strong> de forma algebraica: </p>

$L[y] = Y$<br>
$L[y'] = sY - y(0)$<br>
$L[y''] = s^2Y - sy(0)-y'(0)$<br>
<p>...</p>
<p>Podemos continuar de esta forma evaluando las distintas derivadas</p><br>

<P>Ejemplo: Considere la siguiente ecuacion diferencial linear</p>
$y''-y=e^{-t} , y(0)=1, y'(0)=0$<br>
<p>Podemos aplicar las relaciones anteriores para expresar la ecuacion como:</p>
$s^2Y-s-Y = \frac{1}{s+1}$ <br><br>

<p>Teoremas de Intercambio</p>
<p>Existen 2 teoremas de intercambio, matematicamente establecen lo siguiente</p>
<p>1.- Si  $f(t)$ es multiplicado por $e^{at}$ entonces:</p>
$L[f(t)]e^{at}=F(s-a)$<br>
<p>2.- Se da al intercambiar el eje $t$</p>
<p>A) Queremos desplazar la funcion a $t=a$ si $F(s)=L[f(t)]$ luego:</p>
$u_a(t)f(t-a) = e^{-as}F(s).$ <br>
<p>B) Queremos borrar una parte de la funcion. Luego: </p>
$u_a(t)f(t)=e^{-as}L[f(t+a)].$<br>
<p> Donde la funcion $u_a(t)$, es la funcion de escalon unitario en t=a. Estos teoremas son utiles para trabajar con las funciones a trozos continuas</p><br>

<p>Transformadas de algunas funciones basicas: </p>
$1. L[1]=\frac{1}{s}$<br>
$2. L[t^n]= \frac{n!}{s^{n+1}}, n=1,2,3,...$<br>
$3. L[e^{at}] = \frac{1}{s-a}$<br>
$4. L[e^{at}t^n] = \frac{n!}{(s-a)^{n+1}}$<br>
$5. L[sin (kt)] = \frac{k}{s^2-k^2} $<br>
$6. L[cos (kt)] = \frac{s}{s^2+k^2}$<br>
$7. L[sinh (kt)] = \frac{k}{s^2-k^2}$<br>
$8. L[cosh (kt)] = \frac{s}{s^2-k^2}$<br><br>

<p>Ejemplo: Apliquemos la transformada inversa de Laplace a la siguiente funcion</p>
$F(s) = \frac{1+e^{-πs}}{s^2+1}$<br>
<p> Podemos separar la fraccion en dos terminos, vemos que el termino $e^{-πs}$ posee un desplazamiento de $a=-π$. Por otro lado la primera fraccion la podemos encontrar directamente en las inversas directas.</p>
$F(s)=\frac{1}{s^2+1} + \frac{e^{-πs}}{s^2+1}$

$f(t) = sin(t)$ cuando $0<=t<-π$<br>
$f(t) = sin(t) + sin(t-π)$ cuando $t>=π$<br>

$f(t) = sin(t)$ para $0<=t<π$
